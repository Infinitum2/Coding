```python
class ListNode:
     def __init__(self, val=0, next=None):
         self.val = val
         self.next = next
```

```python
list = ListNode(val = 0, next = ListNode(val = 1, next = None))
current = list
while True:
  current = current.next
  if current == None:
    break
  else:
    print(current.val)
```
